#Removing the duplicates

```python
original_rows = len(df)
df = df.drop_duplicates()
print(f"🗑️  Exact duplicates removed : {original_rows - len(df):,}")
print(f"   Rows remaining         : {len(df):,}")
```

#Cleaning the Date & Time

```python
print("=" * 70)
print(" CLEANING DATE & TIME")
print("=" * 70)

# --- Col 2: Date Of Stop -------------------------------------------------
df['Date Of Stop'] = pd.to_datetime(df['Date Of Stop'], errors='coerce')

future_mask = df['Date Of Stop'] > pd.Timestamp.now()
print(f"📅 Future dates nullified : {future_mask.sum():,}")
df.loc[future_mask, 'Date Of Stop'] = pd.NaT
```

```python
df['Time Of Stop'] = (
    df['Time Of Stop']
    .astype(str)
    .str.replace('.', ':', regex=False)
)
df['Time Of Stop'] = (
    pd.to_datetime(df['Time Of Stop'], format='%H:%M:%S', errors='coerce')
    .dt.time
)

# "Combine with Date Of Stop if you need a full timestamp"
df['DateTime'] = pd.to_datetime(
    df['Date Of Stop'].astype(str) + ' ' + df['Time Of Stop'].astype(str),
    errors='coerce'
)

print(f"✅ Valid dates     : {df['Date Of Stop'].notna().sum():,}")
print(f"   Valid times     : {df['Time Of Stop'].notna().sum():,}")
print(f"   Valid DateTimes : {df['DateTime'].notna().sum():,}")
```

# Cleaning text and Categorical columns


```python
print("=" * 70)
print(" CLEANING TEXT / CATEGORICAL COLUMNS")
print("=" * 70)

# --- Col 4: Agency --------------------------------------------------------
# "Standardize text (uppercase). Only a few unique categories."
df['Agency'] = df['Agency'].astype(str).str.upper().str.strip()
print(f"🏢 Agency unique values : {df['Agency'].nunique()}")
```

```python

# --- Col 5: SubAgency -----------------------------------------------------
# "Standardize spacing & punctuation.

df['SubAgency'] = (
    df['SubAgency'].astype(str)
    .str.strip()
    .str.replace(r'\s+',  ' ',  regex=True)
    .str.replace(r'\s*,\s*', ', ', regex=True)
)
# Extract district number: "3rd District, Silver Spring" → 3
df['District_Number'] = (
    df['SubAgency']
    .str.extract(r'(\d+)(?:st|nd|rd|th)\s+District', expand=False)
)
df['District_Number'] = pd.to_numeric(df['District_Number'], errors='coerce')
print(f"📍 Districts extracted : {df['District_Number'].notna().sum():,}")
```

```python

# --- Col 6: Description ---------------------------------------------------
# "Strip spaces. Standardize case. Categorize using keyword groups
#  (speeding, registration, license, etc.)."
df['Description'] = df['Description'].astype(str).str.strip().str.title()

# Keyword-group categorisation
KEYWORD_GROUPS = [
    ('Speeding',           ['speed', 'speeding']),
    ('Registration',       ['registration']),
    ('License',            ['license']),
    ('Traffic Signal',     ['signal', 'traffic light', 'red light', 'stop sign']),
    ('Seatbelt',           ['seatbelt', 'seat belt', 'belt']),
    ('Insurance',          ['insurance']),
    ('Right of Way',       ['right of way', 'right-of-way']),
    ('Lane Violation',     ['lane']),
    ('Tailgating',         ['tailgating', 'following too close']),
    ('Distracted Driving', ['distracted', 'cell phone', 'texting']),
    ('Alcohol Related',    ['alcohol', 'dui', 'dwi', 'drunk']),
    ('Parking',            ['parking']),
    ('Equipment',          ['equipment', 'defective', 'headlight', 'taillight']),
]


def categorize_violation(description):
    """Map a violation description to its keyword-group category."""
    desc_lower = str(description).lower()
    for category, keywords in KEYWORD_GROUPS:
        if any(kw in desc_lower for kw in keywords):
            return category
    return 'Other'


df['Violation_Category'] = df['Description'].apply(categorize_violation)
print(f"📝 Violation categories:\n{df['Violation_Category'].value_counts().to_string()}")
```

```python
df['Location'] = (
    df['Location'].astype(str)
    .str.strip()
    .str.replace(r'\s*/\s*',  ' / ', regex=True)   # standardise /
    .str.replace(r'\s*@\s*',  ' @ ', regex=True)   # standardise @
    .str.replace(r'\s+', ' ', regex=True)
)
# Extract primary road name (first segment)
df['Road_Name'] = (
    df['Location']
    .str.split(r'\s*/\s*|\s*@\s*', regex=True)
    .str[0]
    .str.strip()
)
print(f"📍 Unique road names : {df['Road_Name'].nunique()}")
```

```python
# "Normalize to uppercase. Validate values (should all be MD)."
df['State'] = df['State'].astype(str).str.upper().str.strip()
non_md = df[df['State'] != 'MD']['State'].value_counts()
print(f"🏛  Non-MD State entries : {non_md.sum():,}")

# --- Col 39: Driver City --------------------------------------------------
# "Standardize city names. Fix typos or formatting inconsistencies."
df['Driver City'] = df['Driver City'].astype(str).str.title().str.strip()

# --- Col 40: Driver State -------------------------------------------------
# "Standardize uppercase. Many should be MD."
df['Driver State'] = df['Driver State'].astype(str).str.upper().str.strip()
print(f"🏛  Top Driver States:\n{df['Driver State'].value_counts().head(5).to_string()}")

# --- Col 41: DL State -----------------------------------------------------
# "Standardize uppercase. Validate codes (should be 2-letter states)."
df['DL State'] = df['DL State'].astype(str).str.upper().str.strip()
invalid_dl = df[df['DL State'].str.len() != 2]['DL State'].value_counts()
print(f"🏛  Invalid DL State codes (not 2 chars) : {invalid_dl.sum():,}")
df.loc[df['DL State'].str.len() != 2, 'DL State'] = np.nan

print("✅ Text / categorical columns cleaned")
```

# Cleaning Geographic Conditions(Latitude, Longitude, Geolocation)

```python
print("=" * 70)
print(" CLEANING GEOGRAPHIC COORDINATES")
print("=" * 70)

# --- Col 8: Latitude ------------------------------------------------------
df['Latitude'] = pd.to_numeric(df['Latitude'], errors='coerce')
zero_lat = (df['Latitude'] == 0).sum()
df.loc[df['Latitude'] == 0, 'Latitude'] = np.nan
# US latitudes: ~24 to ~50
invalid_lat = ((df['Latitude'] < 24) | (df['Latitude'] > 50)).sum()
df.loc[(df['Latitude'] < 24) | (df['Latitude'] > 50), 'Latitude'] = np.nan
print(f"🌍 Latitude  — zeros: {zero_lat:,} | out-of-range: {invalid_lat:,}")
```

```python

# --- Col 9: Longitude -----------------------------------------------------
df['Longitude'] = pd.to_numeric(df['Longitude'], errors='coerce')
zero_lon = (df['Longitude'] == 0).sum()
df.loc[df['Longitude'] == 0, 'Longitude'] = np.nan
invalid_lon = ((df['Longitude'] < -125) | (df['Longitude'] > -65)).sum()
df.loc[(df['Longitude'] < -125) | (df['Longitude'] > -65), 'Longitude'] = np.nan
print(f"🌍 Longitude — zeros: {zero_lon:,} | out-of-range: {invalid_lon:,}")

valid_pairs = df[['Latitude', 'Longitude']].notna().all(axis=1).sum()
print(f"✅ Valid coordinate pairs : {valid_pairs:,} ({valid_pairs/len(df)*100:.1f} %)")
```

```python

# --- Col 43: Geolocation --------------------------------------------------
if 'Geolocation' in df.columns:
    # Extract lat / lon from tuple string  "(38.99, -77.01)"
    geo_lat = pd.to_numeric(
        df['Geolocation'].astype(str).str.extract(r'\(\s*([-\d.]+)', expand=False),
        errors='coerce'
    )
    geo_lon = pd.to_numeric(
        df['Geolocation'].astype(str).str.extract(r',\s*([-\d.]+)\s*\)', expand=False),
        errors='coerce'
    )

    # Replace invalid (0, 0)
    bad = (geo_lat == 0) & (geo_lon == 0)
    geo_lat[bad] = np.nan
    geo_lon[bad] = np.nan

    # Validate against existing Latitude / Longitude
    both_valid = (
        geo_lat.notna() & geo_lon.notna() &
        df['Latitude'].notna() & df['Longitude'].notna()
    )
    matches = (
        (geo_lat[both_valid].round(4) == df.loc[both_valid, 'Latitude'].round(4)) &
        (geo_lon[both_valid].round(4) == df.loc[both_valid, 'Longitude'].round(4))
    ).sum()
    total_both = both_valid.sum()

    if total_both > 0 and (matches / total_both) > 0.9:
        print(f"🗺️  Geolocation redundant ({matches:,}/{total_both:,} match) → dropped")
    else:
        # Not redundant — use it to fill missing Lat/Lon
        df['Latitude']  = df['Latitude'].fillna(geo_lat)
        df['Longitude'] = df['Longitude'].fillna(geo_lon)
        print("🗺️  Geolocation used to fill missing Lat/Lon")

    df = df.drop(columns=['Geolocation'])
else:
    print("⚠️  'Geolocation' column not found — skipping")
```

# Cleaning Boolean Columns

```python
print("=" * 70)
print(" CLEANING BOOLEAN COLUMNS")
print("=" * 70)

BOOLEAN_COLUMNS = [
    'Accident', 'Belts', 'Personal Injury', 'Property Damage',
    'Fatal', 'Commercial License', 'HAZMAT', 'Commercial Vehicle',
    'Alcohol', 'Work Zone', 'Search Conducted', 'Contributed To Accident'
]

for col in BOOLEAN_COLUMNS:
    if col in df.columns:
        df[col] = df[col].astype(str).str.upper().str.strip()
        df[col] = df[col].isin(['YES', 'Y', 'TRUE', '1', 'T'])
        print(f"   ✅ {col:32s} True: {df[col].sum():,}")

print("✅ Boolean columns standardised")
```

#Cleaning Search Columns(Disposition, Outcome, Reason, Reason for Stop, Type, Arrest Reason)

```python

print("=" * 70)
print(" CLEANING SEARCH COLUMNS")
print("=" * 70)

SEARCH_COLUMNS = [
    'Search Disposition', 'Search Outcome', 'Search Type',
    'Search Reason', 'Search Reason For Stop', 'Search Arrest Reason'
]

for col in SEARCH_COLUMNS:
    if col in df.columns:
        df[col] = df[col].fillna('Not Applicable').astype(str).str.strip()
        df.loc[df[col] == '', col] = 'Not Applicable'
        print(f"   🔍 {col}")

# Cols 23, 24, 26 — "Standardize codes" / "Standardize legal code formatting"
LEGAL_CODE_COLS = ['Search Reason', 'Search Reason For Stop', 'Search Arrest Reason']
for col in LEGAL_CODE_COLS:
    if col in df.columns:
        df[col] = (
            df[col].astype(str)
            .str.replace(r'\s+\(', '(', regex=True)   # space before (
            .str.replace(r'\(\s+', '(', regex=True)   # space after (
            .str.replace(r'\s+\)', ')', regex=True)   # space before )
        )

print("✅ Search columns cleaned")
```

# Cleaning Vehicle Columns

```python
print("=" * 70)
print(" CLEANING VEHICLE COLUMNS")
print("=" * 70)

# --- Col 29: Year ---------------------------------------------------------
df['Year'] = pd.to_numeric(df['Year'], errors='coerce')
bad_years = ((df['Year'] < 1960) | (df['Year'] > 2025)).sum()
df.loc[(df['Year'] < 1960) | (df['Year'] > 2025), 'Year'] = np.nan
print(f"📅 Year — invalid removed : {bad_years:,}")

# --- Col 30: Make ---------------------------------------------------------
df['Make'] = df['Make'].astype(str).str.upper().str.strip()

MAKE_ABBREVIATIONS = {
    'CHEV': 'CHEVROLET', 'CHEVY': 'CHEVROLET',
    'TOYT': 'TOYOTA',    'HOND': 'HONDA',
    'NISS': 'NISSAN',    'DODG': 'DODGE',
    'HYUN': 'HYUNDAI',   'MERZ': 'MERCEDES-BENZ',
    'BENZ': 'MERCEDES-BENZ',
    'VOLKS': 'VOLKSWAGEN', 'VW': 'VOLKSWAGEN',
    'LEXS': 'LEXUS',     'INFI': 'INFINITI',
    'ACUR': 'ACURA',     'SUBA': 'SUBARU',
    'MAZD': 'MAZDA',     'CHRY': 'CHRYSLER',
    'BUIC': 'BUICK',     'CADI': 'CADILLAC',
    'LINC': 'LINCOLN',   'PONT': 'PONTIAC',
}
df['Make'] = df['Make'].replace(MAKE_ABBREVIATIONS)
print(f"🚗 Make — unique after standardisation : {df['Make'].nunique()}")

# --- Col 31: Model --------------------------------------------------------
df['Model'] = df['Model'].astype(str).str.upper().str.strip()

# --- Col 32: Color --------------------------------------------------------
df['Color'] = df['Color'].astype(str).str.upper().str.strip()

COLOR_STANDARD_MAP = {
    'BLK': 'BLACK',  'WHI': 'WHITE',  'WHT': 'WHITE',
    'SIL': 'SILVER', 'GRY': 'GRAY',   'GRA': 'GRAY',
    'BLU': 'BLUE',   'GRN': 'GREEN',  'YEL': 'YELLOW',
    'TAN': 'TAN',    'BRO': 'BROWN',  'BRN': 'BROWN',
    'GLD': 'GOLD',   'MAR': 'MAROON', 'ORA': 'ORANGE',
    'PUR': 'PURPLE', 'PNK': 'PINK',   'BGE': 'BEIGE',
}
df['Color'] = df['Color'].replace(COLOR_STANDARD_MAP)
print(f"🎨 Color — unique after standardisation : {df['Color'].nunique()}")

# --- Col 28: VehicleType --------------------------------------------------
# "Split into Code + Category if useful. Standardize spaces around hyphens."
df['VehicleType'] = df['VehicleType'].astype(str).str.strip()
df['VehicleType_Code']     = df['VehicleType'].str.extract(r'^(\d+)', expand=False)
df['VehicleType_Category'] = (
    df['VehicleType']
    .str.replace(r'^\d+\s*[-–]\s*', '', regex=True)
    .str.strip()
)
print(f"🚘 VehicleType categories:\n{df['VehicleType_Category'].value_counts().head(6).to_string()}")

print("✅ Vehicle columns cleaned")
```

# Cleaning Violation and Arrest Columns

```python
print("=" * 70)
print(" CLEANING VIOLATION & ARREST COLUMNS")
print("=" * 70)

# --- Col 33: Violation Type -----------------------------------------------
df['Violation Type'] = df['Violation Type'].astype(str).str.title().str.strip()
print(f"⚖️  Violation Type values:\n{df['Violation Type'].value_counts().to_string()}")

# --- Col 34: Charge -------------------------------------------------------

df['Charge'] = df['Charge'].astype(str).str.strip()
charge_per_stop = (
    df.groupby('SeqID')['Charge']
    .nunique()
    .rename('Charge_Count')
    .reset_index()
)
df = df.merge(charge_per_stop, on='SeqID', how='left')
multi_charge = (charge_per_stop['Charge_Count'] > 1).sum()
print(f"⚖️  Stops with multiple charges : {multi_charge:,}")

# --- Col 35: Article ------------------------------------------------------

df['Article'] = df['Article'].astype(str).str.strip()
df['Article'] = (
    df.groupby('SeqID')['Article']
    .transform(
        lambda series: series.fillna(
            series.mode().iloc[0] if not series.mode().empty else 'Unknown'
        )
    )
)
print(f"📘 Article — unique : {df['Article'].nunique()}")
```

```python
df['Arrest Type'] = df['Arrest Type'].astype(str).str.strip()
df['Arrest_Type_Code'] = (
    df['Arrest Type']
    .str.extract(r'^([A-Za-z])\s*[-–]', expand=False)
    .str.upper()
)
df['Arrest_Type_Desc'] = (
    df['Arrest Type']
    .str.replace(r'^[A-Za-z]\s*[-–]\s*', '', regex=True)
    .str.strip()
)
print(f"🚓 Arrest Type codes:\n{df['Arrest_Type_Code'].value_counts().to_string()}")

print("✅ Violation & Arrest columns cleaned")
```

# Cleaning Geographic Columns

```python

print("=" * 70)
print(" CLEANING DEMOGRAPHIC COLUMNS")
print("=" * 70)

# --- Col 37: Race ---------------------------------------------------------
df['Race'] = df['Race'].astype(str).str.upper().str.strip()
# Do NOT infer / fill missing — per PDF
print(f"🧬 Race:\n{df['Race'].value_counts().to_string()}")

# --- Col 38: Gender -------------------------------------------------------
df['Gender'] = df['Gender'].astype(str).str.upper().str.strip()
df.loc[~df['Gender'].isin(['M', 'F']), 'Gender'] = 'UNKNOWN'
print(f"\n👤 Gender:\n{df['Gender'].value_counts().to_string()}")

print("✅ Demographic columns cleaned")
```

# Feature Engineering

```python

print("=" * 70)
print(" FEATURE ENGINEERING")
print("=" * 70)

# --- Temporal features ----------------------------------------------------
df['Hour']       = df['DateTime'].dt.hour
df['DayOfWeek']  = df['DateTime'].dt.day_name()
df['Month']      = df['DateTime'].dt.month_name()
df['Year_Stop']  = df['DateTime'].dt.year
df['DayOfMonth'] = df['DateTime'].dt.day
df['WeekOfYear'] = df['DateTime'].dt.isocalendar().week.astype('Int64')

# --- Time-of-day buckets --------------------------------------------------
df['Time_Bucket'] = pd.cut(
    df['Hour'],
    bins=[-1, 6, 12, 18, 24],
    labels=[
        'Night (12AM-6AM)',
        'Morning (6AM-12PM)',
        'Afternoon (12PM-6PM)',
        'Evening (6PM-12AM)'
    ]
).astype(str)
df.loc[df['Hour'].isna(), 'Time_Bucket'] = 'Unknown'
print(f"🕐 Time Buckets:\n{df['Time_Bucket'].value_counts().to_string()}")

# --- Weekend flag ---------------------------------------------------------
df['Is_Weekend'] = df['DayOfWeek'].isin(['Saturday', 'Sunday'])

# --- Accident severity indicator (High_Risk) ------------------------------
# Combines Accident, Personal Injury, Fatal, Alcohol into one flag
df['High_Risk'] = (
    df['Accident'] |
    df['Personal Injury'] |
    df['Fatal'] |
    df['Alcohol']
)
print(f"\n⚠️  High-Risk violations : {df['High_Risk'].sum():,} "
      f"({df['High_Risk'].mean()*100:.2f} %)")

# --- Violation count per stop ---------------------------------------------
viol_counts = (
    df.groupby('SeqID').size()
    .rename('Violation_Count')
    .reset_index()
)
df = df.merge(viol_counts, on='SeqID', how='left')
print(f"📊 Max violations in one stop : {df['Violation_Count'].max()}")

# --- Vehicle age ----------------------------------------------------------
df['Vehicle_Age'] = df['Year_Stop'] - df['Year']
df.loc[df['Vehicle_Age'] < 0, 'Vehicle_Age'] = np.nan
print(f"🚗 Median vehicle age : {df['Vehicle_Age'].median():.0f} years")

print("✅ Feature engineering complete")
```

# Optimise Data Types

```python

print("=" * 70)
print(" OPTIMISING DATA TYPES")
print("=" * 70)

mem_before = df.memory_usage(deep=True).sum() / 1024**2
print(f"Memory before : {mem_before:.2f} MB")

# Low-cardinality object → category
for col in df.select_dtypes(include=['object']).columns:
    if df[col].nunique() / len(df) < 0.5:
        df[col] = df[col].astype('category')

# Downcast int & float
for col in df.select_dtypes(include=['int64', 'int32']).columns:
    df[col] = pd.to_numeric(df[col], downcast='integer')
for col in df.select_dtypes(include=['float64', 'float32']).columns:
    df[col] = pd.to_numeric(df[col], downcast='float')

mem_after = df.memory_usage(deep=True).sum() / 1024**2
print(f"Memory after  : {mem_after:.2f} MB  (↓ {(1 - mem_after/mem_before)*100:.1f} %)")
```
